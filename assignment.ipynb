```python
import struct
import pandas as pd
```

```python
schema_file="Schema.xlsx"
schema_df=pd.read_excel(schema_file)
```

```python
schema_df.head()
```

```python
binary_file="R520000320240201170000_2024020117050024632.bin"
#Read the binary file
with open(binary_file,'rb') as f:
    binary_data =f.read()
```

```python
binary_data
```

```python
#Extracting and matching the file
parsed_data={}
for index,row in schema_df.iterrows():
    field_name=row["Column"]
    start_index=row['Starting Index']
    field_len=row['Len']
    #slicing the binary data
    field_bytes=binary_data[start_index:start_index+field_len]
    #Decode the bytes
    try:
        decoded_value=int(field_bytes.decode('utf-8').strip())
    except ValueError:
        decoded_value=field_bytes.decode('utf-8').strip()
    #save the parsed value
    parsed_data[field_name]=decoded_value
print("Parsed Data:")
for key ,value in parsed_data.items():
    print(f'{key}:{value}')
```

```python
parsed_df=pd.DataFrame([parsed_data])
parsed_df.head()
```

```python
outputfile ="output.csv"
parsed_df.to_csv(outputfile,index=False,header=True)
print(f"Parsed data has been saved to {outputfile}")
```
